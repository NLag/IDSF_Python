```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

filename='RF_result14-04-2023_0108.csv'
df = pd.read_csv(filename,na_values='',names=['label','predict'], header=None)
df.head()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score, f1_score,classification_report

y_true = df['label']
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15,10))
models = ['DT','LSVC','Neural','RF','SGDC','XGBoost']
n = len(df)

for name,ax in zip(models,axes.flatten()) :
    y_pred = df[name]
    print(name)
    emulation_accuracy = accuracy_score(y_true, y_pred)
    print('test acc: {:.7f}'.format(emulation_accuracy))
    emulation_precision = precision_score(y_true, y_pred)
    print('test precision: {:.8f}'.format(emulation_precision))
    emulation_recall = recall_score(y_true, y_pred)
    print('test recall: {:.8f}'.format(emulation_recall))
    emulation_f1 = f1_score(y_true, y_pred)
    print('test F1: {:.8f}'.format(emulation_f1))

    # print(accuracy_score(y_true, y_pred, normalize=False))
    # print(classification_report(y_true, y_pred))

    cm = confusion_matrix(y_true,y_pred)
    tn, fp, fn, tp = confusion_matrix(y_true,y_pred).ravel()
    print('FNR: {:.8f}'.format(fn/(fn+tp)))
    print('FPR: {:.8f}'.format(fp/(fp+tn)))

    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(ax=ax,values_format='')
    disp.ax_.set_title(name)

plt.tight_layout()
plt.show()
```
